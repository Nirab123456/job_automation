```python
from io import StringIO

# Input text data
file_path = 'rawData/ON-08-05-2024.txt'
with open(file_path, 'r') as file:
    data = file.readlines()

# Define the expected number of fields
expected_num_fields = 8

# Filter out lines with more than the expected number of fields
filtered_data = [line for line in data if len(line.split('|')) == expected_num_fields]

# Combine the filtered lines back into a single string
filtered_data_str = ''.join(filtered_data)

# Use StringIO to read the string data as if it were a file
data_io = StringIO(filtered_data_str)

# Read the data into a DataFrame
df = pd.read_csv(data_io, delimiter='|')

# Save the DataFrame to a CSV file
df.to_csv('Toronto-08.csv', index=False)

print("Data has been successfully converted and saved to Toronto-08.csv")
```

```python
from datetime import datetime, timedelta
```

```python
# Convert 'Post Date' to datetime
df['Post Date'] = pd.to_datetime(df['Post Date'])

# Define the current date (you can use datetime.now() for the current date)
current_date = datetime.now()

# Calculate the date 15 days ago
date_15_days_ago = current_date - timedelta(days=15)

# Filter the DataFrame to keep rows where 'Post Date' is within the last 15 days
filtered_df = df[df['Post Date'] >= date_15_days_ago]

# Display the filtered DataFrame
print(filtered_df)
```
